## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
players[1]
````

````python
# Example output:
#
# Ash's power is 57.
# 57
# Brock's power is 119.
# 119

def calc_power(player_dict, pokedex, player_id):
    total_attack = 0
    total_defense = 0
    for pokemon_id, pokemon_dict in player_dict[player_id]['player_pokemon'].items():
        total_attack += pokemon_dict['attack']
        total_defense += pokemon_dict['defense']
    power = total_attack + total_defense
    print '{}\'s power is {}.'.format(player_dict[player_id]['player_name'],power)
        
````

````python
# Example output:
#
# {'Attack': 30.0,
#  'Defense': 50.0,
#  'HP': 40.0,
#  'Name': 'Voltorb',
#  'SpecialAttack': 55.0,
#  'SpecialDefense': 55.0,
#  'Speed': 100.0,
#  'Total': 330.0,
#  'Type': 'Electric'}


def process_pokedex(parsed_pokedex):
    pokedex_dict = {}
    
    for i,lst in enumerate(parsed_pokedex):
        if i == 0: # skip first row (column headers)
            pass
        
        pokedex_number = lst[0]
        name = lst[1]
        pokemon_type = lst[2]
        total = lst[3]
        hp = lst[4]
        attack = lst[5]
        defense = lst[6]
        special_attack = lst[7]
        special_defense = lst[7]
        speed = lst[8]
        
        pokedex_dict[pokedex_number] = {
                            'Attack': attack,
                            'Defense': defense,
                            'HP': hp,
                            'Name': name,
                            'SpecialDefense': special_defense,
                            'SpecialAttack': special_attack,
                            'Speed': speed,
                            'Type': pokemon_type,
                            'Total': total
                            }
    
    return pokedex_dict
    
````
